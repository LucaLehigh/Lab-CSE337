# Gaussian Process

```python
import torch
import gpytorch
import numpy as np

# ============================================================
# ---- Base Exact GP model -------------------
# ============================================================

class ExactGPModel(gpytorch.models.ExactGP):
    def __init__(self, train_x, train_y, likelihood, kernel='RBF', ard_dims=None):
        super().__init__(train_x, train_y, likelihood)
        self.mean_module = gpytorch.means.ConstantMean()

        # Choose kernel type
        if kernel == 'RBF':
            base_kernel = gpytorch.kernels.RBFKernel(ard_num_dims=ard_dims)
        elif kernel == 'Matern':
            base_kernel = gpytorch.kernels.MaternKernel(nu=2.5, ard_num_dims=ard_dims)
        elif kernel == 'RQ':
            base_kernel = gpytorch.kernels.RationalQuadraticKernel(ard_num_dims=ard_dims)
        else:
            raise ValueError(f"Unsupported kernel type: {kernel}")

        self.covar_module = gpytorch.kernels.ScaleKernel(base_kernel)

    def forward(self, x):
        mean_x = self.mean_module(x)
        covar_x = self.covar_module(x)
        return gpytorch.distributions.MultivariateNormal(mean_x, covar_x)


# ============================================================
# ---- Individual GP Manager: handles data + normalization ----
# ============================================================

class GPManager:
    def __init__(self, kernel='RBF', lr=0.03, iters=1000):
        self.kernel = kernel
        self.lr = lr
        self.iters = iters
        self.trained = False
        self.X_train = []
        self.Y_train = []

        self.likelihood = None
        self.model = None


    # ----------------------------- #
    #        FIT / INITIAL TRAIN    #
    # ----------------------------- #
    def fit(self, X, Y):
        X = torch.tensor(X, dtype=torch.float32)
        Y = torch.tensor(Y, dtype=torch.float32)

        self.X_train = X.clone()
        self.Y_train = Y.clone()

        self.retrain()

    def retrain(self):
        # Normalize the input/output and then train the model
        self._compute_normalization()
        self._train_model()

    # ----------------------------- #
    #       ADD NEW DATA POINTS     #
    # ----------------------------- #
    def add_data(self, X_new, Y_new, retrain=True):
        """
        Add new data (single or batch) to GP training set.
        Optionally retrain the model.
        """
        X_new = torch.tensor(X_new, dtype=torch.float32)
        Y_new = torch.tensor(Y_new, dtype=torch.float32).flatten()  # ensure 1D

        # Also flatten stored Y to ensure shape consistency
        if self.Y_train.ndim > 1:
            self.Y_train = self.Y_train.flatten()

        # Append new data
        self.X_train = torch.cat([self.X_train, X_new], dim=0)
        self.Y_train = torch.cat([self.Y_train, Y_new], dim=0)

        if retrain:
            self.retrain()

    # ----------------------------- #
    #         INTERNAL UTILS        #
    # ----------------------------- #
    def _compute_normalization(self):
        self.X_mean, self.X_std = self.X_train.mean(0), self.X_train.std(0)
        self.Y_mean, self.Y_std = float(self.Y_train.mean(0)), float(self.Y_train.std(0))

        Xn = (self.X_train - self.X_mean) / self.X_std
        Yn = (self.Y_train - self.Y_mean) / self.Y_std

        self.Xn, self.Yn = Xn, Yn

    def dataset(self):
        # Extract training data (already stored inside the wrapper)
        X_train = self.X_train.numpy()
        Y_train = self.Y_train.numpy()
        return X_train, Y_train


    def _train_model(self):
        self.likelihood = gpytorch.likelihoods.GaussianLikelihood()
        self.likelihood.noise_covar.initialize(noise=1e-3)
        self.model = ExactGPModel(self.Xn, self.Yn, self.likelihood,
                                  kernel=self.kernel, ard_dims=self.X_train.shape[-1])

        # -------------------------------------------------
        # Train the model
        # -------------------------------------------------
        self.train_gp(self.model, self.likelihood, self.Xn, self.Yn)
        self.trained = True

    def train_gp(self, model, likelihood, x, y):
        model.train(); likelihood.train()
        opt = torch.optim.Adam(model.parameters(), lr=self.lr)
        mll = gpytorch.mlls.ExactMarginalLogLikelihood(likelihood, model)
        for _ in range(self.iters):
            opt.zero_grad()
            out = model(x)
            loss = -mll(out, y)
            loss.backward()
            opt.step()
        model.eval(); likelihood.eval()

    def predict(self, X):
        if not self.trained:
            raise RuntimeError("GP has not been trained yet.")

        # Normalize the input
        X = torch.as_tensor(X, dtype=torch.float32)
        Xn = (X - self.X_mean) / self.X_std
        with torch.no_grad(), gpytorch.settings.fast_pred_var():
            pred = self.likelihood(self.model(Xn))

            # Unonormalize the output
            mean = pred.mean.cpu().numpy() * self.Y_std + self.Y_mean
            var = pred.variance.cpu().numpy() * (self.Y_std ** 2)
        # Return the unnormalized output
        return mean, var
```

# Setup environment and collect initial data

```python
import gymnasium as gym
# import gym
import numpy as np
import torch
import matplotlib.pyplot as plt

# --------------------------------------------------------------
# 1.  Environment setup
# --------------------------------------------------------------
env = gym.make("MountainCarContinuous-v0", max_episode_steps=500)
n_state = env.observation_space.shape[0]    # [position, velocity]
n_action = env.action_space.shape[0]        # 1 continuous action
n_output = 2                                # Δpos, Δvel


# --------------------------------------------------------------
# 2.  Collect initial random data
# --------------------------------------------------------------
data = {'X': [], 'Y': []}

obs, _ = env.reset(seed=0)
for _ in range(200):  # collect 500 transitions
    a = env.action_space.sample()  # random continuous action in [-1,1]
    x_next, r, term, trunc, _ = env.step(a)
    data['X'].append(np.concatenate([obs, a]))   # [pos, vel, action]
    data['Y'].append(x_next - obs)               # Δstate
    obs = x_next
    if term or trunc:
        obs, _ = env.reset()

X = np.array(data['X'])[::3]
Y = np.array(data['Y'])[::3]
print(f"Collected {len(X)} transitions.")


# --------------------------------------------------------------
# 3.  Train GP dynamics models
# --------------------------------------------------------------
gps = [GPManager(kernel='RBF', iters=300) for _ in range(n_output)]

for d in range(n_output):
    gps[d].fit(X, Y[:, d])
    print(f"Trained GP for Δstate[{d}] with {len(X)} samples.")
```

# MPC


```python
import numpy as np
import matplotlib.pyplot as plt

# ============================================================
#  Cost function for Mountain Car (simple shaping)
# ============================================================
# Try to get as far to the right (goal) as possible, with small action/velocity penalty
goal_pos = env.unwrapped.goal_position if hasattr(env.unwrapped, "goal_position") else 0.45

def cost_fn(s, a):
    pos, vel = s
    pos_cost = (goal_pos - pos) ** 2         # want pos → goal_pos
    vel_cost = 0.1 * (vel ** 2)             # discourage crazy velocities
    act_cost = 0.001 * (a ** 2)             # tiny penalty on control effort
    return pos_cost + vel_cost + act_cost


# ============================================================
#  Rollout using the GP dynamics model
# ============================================================
def roll_out_trajectory(s0, Ui, gamma=1.0):
    """
    Roll out a single action sequence Ui starting from state s0
    using the learned GP dynamics (gps[0] for Δpos, gps[1] for Δvel).
    Returns:
        traj: [H+1, 2] array of states
        total_cost: scalar
    """
    s = s0.copy()                     # current state [pos, vel]
    traj = [s0.copy()]
    H = len(Ui)
    discount = 1.0
    total_cost = 0.0

    for k in range(H):
        a = Ui[k]                     # scalar action in [-1, 1]

        # --------------------------
        # Predict Δstate with GPs
        # --------------------------
        x_in = np.array([[s[0], s[1], a]], dtype=np.float32)  # [1, 3]

        dpos_mean, _ = gps[0].predict(x_in)   # shape (1,)
        dvel_mean, _ = gps[1].predict(x_in)   # shape (1,)

        delta = np.array([dpos_mean[0], dvel_mean[0]])
        s = s + delta                         # next state

        traj.append(s.copy())

        # accumulate discounted cost
        total_cost += cost_fn(s, a) * discount
        discount *= gamma

    return np.array(traj), total_cost


def roll_out_n_trajectories(state, U, gamma=1.0):
    """
    Roll out all sampled action sequences in U from the same initial state.
    U has shape (N_seq, H).
    Returns:
        all_traj: list of trajectories
        J: [N_seq] array of total costs
    """
    N_seq = len(U)
    J = np.zeros(N_seq)
    all_traj = []

    for i in range(N_seq):
        traj, total_cost = roll_out_trajectory(state, U[i], gamma=gamma)
        J[i] = total_cost
        all_traj.append(traj)

    return all_traj, J


# Optional: simple plotting helper for predicted trajectories
def plot_trajectories(s0, traj_list):
    """
    Plot position vs time for a list of predicted trajectories.
    """
    plt.figure(figsize=(8, 4))
    for traj in traj_list:
        plt.plot(traj[:, 0], alpha=0.5)   # plot position only
    plt.axhline(goal_pos, color='r', linestyle='--', label='goal')
    plt.xlabel("Time (model steps)")
    plt.ylabel("Position")
    plt.title("Predicted trajectories (position vs time)")
    plt.legend()
    plt.tight_layout()
    plt.show()


# ============================================================
#  MPC controller (Random Shooting / MPPI-Simplified)
# ============================================================
def mpc_action(state, H=25, N_seq=200, gamma=1.0,
               plot=False, plot_best=10):
    """
    Random Shooting MPC (MPPI-simplified) using GP dynamics.

    1. Sample N_seq random action sequences of length H.
    2. Roll them out with the GP model.
    3. Choose the sequence with lowest cost.
    4. Return the FIRST action of that sequence.
    """
    # 1. Sample action sequences in [-1, 1]
    U = np.random.uniform(-1.0, 1.0, size=(N_seq, H))

    # 2. Roll out and compute costs
    all_traj, J = roll_out_n_trajectories(state, U, gamma=gamma)

    # 3. Get best sequences (by cost)
    best_idx = np.argsort(J)[:plot_best]
    best_seq_idx = np.argmin(J)

    # Optional plotting of top trajectories
    if plot:
        best_traj = [all_traj[i] for i in best_idx]
        plot_trajectories(state, best_traj)

    # 4. Return first action from best sequence
    best_action = U[best_seq_idx, 0]
    return best_action


# ============================================================
#  Run one episode in the real environment with MPC
# ============================================================
H = 25        # MPC horizon
N_seq = 300   # number of sampled sequences
steps = 500   # max episode length
gamma = 1.0   # no discount (you can try 0.99 too)

obs, _ = env.reset(seed=1)
returns = 0.0

trajectory_states = []
trajectory_actions = []

for t in range(steps):
    # 1. Compute MPC action using GP model
    a = mpc_action(obs, H=H, N_seq=N_seq, gamma=gamma, plot=False)

    # 2. Step the true environment
    next_obs, r, term, trunc, _ = env.step([a])
    returns += r

    trajectory_states.append(obs.copy())
    trajectory_actions.append(a)

    # Optional: slow-ish debug printing
    if t % 20 == 0:
        print(f"t={t}, state={obs}, a={a:.3f}, r={r:.3f}")

    obs = next_obs
    if term or trunc:
        print("Episode finished at step", t)
        break

print(f"Episode return: {returns:.2f}")
env.close()

# ============================================================
#  Plot: car position vs time (deliverable)
# ============================================================
trajectory_states = np.array(trajectory_states)  # [T, 2]
positions = trajectory_states[:, 0]

plt.figure(figsize=(8, 4))
plt.plot(positions, marker='o', markersize=2)
plt.axhline(goal_pos, color='r', linestyle='--', label='goal position')
plt.xlabel("Time step")
plt.ylabel("Position")
plt.title("MountainCar position vs time under GP-MPC")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
